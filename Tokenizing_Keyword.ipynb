```python
# preprocessed 된 csv를 모두 병합하는 과정
df_art = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/art_cleaned.csv', delimiter=',')
df_cook = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/cook_cleaned.csv', delimiter=',')
df_computer = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/computer_cleaned.csv', delimiter=',')
df_development = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/development_cleaned.csv', delimiter=',')
df_economy = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/economy_cleaned.csv', delimiter=',')
df_essay = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/essay_cleaned.csv', delimiter=',')
df_health = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/health_cleaned.csv', delimiter=',')
df_history = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/history_cleaned.csv', delimiter=',')
df_hobby = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/hobby_cleaned.csv', delimiter=',')
df_home = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/home_cleaned.csv', delimiter=',')
df_humanity = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/humanity_cleaned.csv', delimiter=',')
df_job = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/job_cleaned.csv', delimiter=',')
df_language = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/language_cleaned.csv', delimiter=',')
df_novel = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/novel_cleaned.csv', delimiter=',')
df_politics = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/politics_cleaned.csv', delimiter=',')
df_reference = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/reference_cleaned.csv', delimiter=',')
df_science = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/science_cleaned.csv', delimiter=',')
df_technology = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/technology_cleaned.csv', delimiter=',')
df_travel = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/preprocessed_data/travel_cleaned.csv', delimiter=',')


df_all = pd.concat([df_art, df_cook, df_computer, df_development, df_economy,
                    df_essay, df_health, df_history, df_hobby, df_home,
                    df_humanity, df_job, df_language, df_novel, df_politics,
                    df_reference, df_science, df_technology, df_travel])

df_all.reset_index(inplace=True, drop=True)

df_all = df_all[["title", "score", "review_num", "keyword", "category"]]
```

```python
# concat 된 파일에서 keyword를 추출
df_keyword = df_all["keyword"]

# csv에 넣을 list들로 만들어줄 2차원 리스트 구현
tokenized_keyword = []

for i in range(len(df_keyword)):

  # csv로 불러왔기 때문에 list의 형식이 아닌 list가 str로 바뀐 형식이다.
  # 따라서 str 형식을 list로 바꿔주는 logic 필요
  x = df_keyword[i]
  x = ast.literal_eval(x)
  new_list = []

  # list 안의 keyword에 대해서 tokenizing 실행
  for j in range(len(x)):
    words = x[j]
    temp_words = mecab.morphs(words)

    # tokenized 된 데이터를 list에 새로 추가
    new_list.append(temp_words)
  
  answer = []
  for element in new_list:
      answer += element

  my_set = set(answer) #집합set으로 변환
  my_list = list(my_set) #list로 변환
  
  tokenized_keyword.append(my_list)
```

```python
# tokenized 된 keyword set들을 새로운 column 'tokenized_keyword' 로 저장하는 과정
df_all['tokenized_keyword'] = tokenized_keyword

df_all.head()

# book_data.csv로 저장!
df_all.to_csv("/content/drive/Shareddrives/데이터마이닝/book_data.csv", mode='w')
```
