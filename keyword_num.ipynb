```python
df = pd.read_csv('/content/drive/Shareddrives/데이터마이닝/book_data.csv')

# len(df.keyword[0].replace("'", "").replace("[", "").replace("]", "").split(","))

# df.keyword

# df.keyword[0].split(',')

keyword_counts = []

for key in df.keyword:
  for keyword in key.replace("'", "").replace("[", "").replace("]", "").split(","):
    keyword_counts.append(keyword)

list = pd.Series(keyword_counts).value_counts().to_dict()
print(list)

wc = WordCloud(background_color="white", max_words=100, width=1000, height=800)
wc = wc.generate_from_frequencies(list)
plt.figure(figsize=(15, 15))
plt.imshow(wc)
plt.axis('off')
plt.show()



# series_to_list = pd.DataFrame(list, columns=['number'])
# print(series_to_list)

# series_to_list.to_csv("/content/drive/Shareddrives/데이터마이닝/keyword_num.csv")


# list = keyword_counts.value_counts().head(20)
# print(list)
```
